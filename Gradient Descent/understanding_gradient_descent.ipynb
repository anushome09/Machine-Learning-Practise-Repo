```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
# Creating a data that follows the linear regression
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13) 
```

```python
# Plotting the data into a graph
import matplotlib.pyplot as plt

plt.scatter(X,y)
```

Using the OLS method, we will find the value of coefficient

```python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()

reg_model.fit(X,y)
```

```python
# Value for the coefficients
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X,reg_model.predict(X), color='blue')
```

Using the Gradient Descent, we will try to reach optimal intercept (paramter) value - 26.15

```python
# Assuming the Slope(coefficient) is constant = 78.35
# For the intercept lets randomly start from b = 0

# Let's get prediction using formula
y_pred = ((78.35 * X) + 0).reshape(4)
y_pred 
```

```python
# Plotting the regression line for the new predicted value
plt.scatter(X, y)
plt.plot(X,reg_model.predict(X), color='red', label='OLS')

plt.plot(X, y_pred, color='green', label='b = 0')
plt.legend()
plt.show()
```

```python
# Applying Gradient descent, by finding the slop and new value for b

m = 78.53
b = 0

loss_slope = -2 * np.sum(y - m* X.ravel() - b)
loss_slope
```

```python
# Applying the learning rate to the slop and find the new value for b
learning_rate = 0.1

b = b - learning_rate * loss_slope
b
```

```python
# Again the same process 
y_pred_new = ((78.35 * X) + b).reshape(4)


# Plotting the regression line for the new predicted value
plt.scatter(X, y)
plt.plot(X,reg_model.predict(X), color='red', label='OLS')

plt.plot(X, y_pred, color='green', label='b = 0')

plt.plot(X, y_pred_new, color='orange', label='b = 20.91')
plt.legend()
plt.show()
```

```python
# Repeating the same
loss_slope = -2 * np.sum(y - m* X.ravel() - b)
b = b - learning_rate * loss_slope

b
```

```python
y_pred_new2 = ((78.35 * X) + b).reshape(4)


# Plotting the regression line for the new predicted value
plt.scatter(X, y)
plt.plot(X,reg_model.predict(X), color='red', label='OLS')

plt.plot(X, y_pred, color='green', label='b = 0')

plt.plot(X, y_pred_new, color='orange', label='b = 20.91')

plt.plot(X, y_pred_new2, color='blue', label='b = 25.09')
plt.legend()
plt.show()
```

```python
# Repeating the same
loss_slope = -2 * np.sum(y - m* X.ravel() - b)
b = b - learning_rate * loss_slope

b
```

```python
y_pred_new3 = ((78.35 * X) + b).reshape(4)


# Plotting the regression line for the new predicted value
plt.scatter(X, y)
plt.plot(X,reg_model.predict(X), color='red', label='OLS')

plt.plot(X, y_pred, color='green', label='b = 0')

plt.plot(X, y_pred_new, color='orange', label='b = 20.91')

plt.plot(X, y_pred_new2, color='blue', label='b = 25.09')

plt.plot(X, y_pred_new3, color='yellow', label='b = 25.09')
plt.legend()
plt.show()
```
