```python
# Creating a class to generate the value of intercept (b)
class GDRegressor:
    
    def __init__(self, learning_rate, epoch):
        self.m = 100
        self.b = -120
        self.learning_rate = learning_rate
        self.epoch = epoch
        
    def fit(self, X, y):
        # calculate the b and m using gradient descent
        for i in range(self.epoch):
            # For intercept b
            loss_slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - (self.learning_rate * loss_slope_b)
            
            # For coefficient m
            loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.b) * X.ravel())
            self.m = self.m - (self.learning_rate * loss_slope_m)
            
            
        print('Intercept b: ', self.b)
        print('Coefficient m: ', self.m)
               
```

```python
# Creating an object of the class
gd = GDRegressor(0.001, 50)
```

```python
gd.fit(X,y)
```
